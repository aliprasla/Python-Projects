```python
from pandas import Series,DataFrame
import pandas as pd
%pylab inline
```

```python
#Unique name for each restaurant
df = pd.read_csv('NYC_Restaurants.csv'
                 ,parse_dates = ['INSPECTION DATE','GRADE DATE','RECORD DATE'])
```

```python
#Q1

df['ZIPCODE'] = df['ZIPCODE'].apply(str)

df['RESTAURANT'] = df['DBA'] + " " + df['BUILDING'] + " " + df['STREET'] + " " + df['ZIPCODE'] 
df['RESTAURANT'][:5]
```

```python
#Q2 -> Num Restaurants -> takes in account duplicates
len(df['RESTAURANT'].value_counts())
```

```python
#Q3 -> Num Chains
chains = df['DBA'].value_counts()[df['DBA'].value_counts() > 1]
len(chains)
```

```python
chains[:20].plot(kind  = "bar")
```

```python
#Q5 what fraction of all restaurants are chains
numRestaurants = len(df)
sum(chains.values)/float(numRestaurants)
```

```python
#Q6 => number nonChain restaurants per Boro
nonChains = df['DBA'].value_counts()[df['DBA'].value_counts() == 1]
nonChainsMask = df['DBA'].isin(nonChains.index)
missingMask = df['BORO'] != "Missing"
df[nonChainsMask & missingMask]['BORO'].value_counts().plot(kind = "bar")
```

```python
#Q7 => fraction of non-chain restaurants per Boro
totalRestPerBoro = df['BORO'][df['BORO'] != "Missing"].value_counts()

(df[nonChainsMask & missingMask]['BORO'].value_counts()/totalRestPerBoro).plot(kind = "bar",title = "Fraction Non Chain Restaurants per Boro")
```

While Manhattan as the most independent restaurants, it does not have the highest fraction of independent restaurants. That title belongs to Brooklyn.

```python
#Q8 => Popularity of Cuisines
df.drop_duplicates(subset = "RESTAURANT")['CUISINE DESCRIPTION'].value_counts()[:20].plot(kind = "bar")
```

```python
#Q9 => Cusines Among Restaurants that never got violation
noViolationMask = pd.isnull(df['VIOLATION CODE'])
df[noViolationMask]['CUISINE DESCRIPTION'].value_counts()[:20].plot(kind = "bar", title = "Top 20 Cusinies of Restaurants without Violations")
```

```python
#Q10 => Cleanest Cuisines

topCuisine = df.drop_duplicates(subset = "RESTAURANT")['CUISINE DESCRIPTION'].value_counts()
topCuisineRestaurantViolations = df[df['CUISINE DESCRIPTION'].isin(topCuisine.index)]


codeMissingMask = pd.isnull(df['VIOLATION CODE'])

noViolationsbyCuisine = topCuisineRestaurantViolations[codeMissingMask]['CUISINE DESCRIPTION'].value_counts()


totalViolationsbyCuisine = topCuisineRestaurantViolations[~codeMissingMask]['CUISINE DESCRIPTION'].value_counts()

(noViolationsbyCuisine/totalViolationsbyCuisine).sort_values(ascending = False)[:10].plot(kind = "bar",title = "Most likely to be clean")
```

```python
#Q11 => common violations in each boro
M1 = ~pd.isnull(df['VIOLATION CODE'])
M2 = df['BORO'] == 'Missing'
filtered = df[M1 & ~M2 ]

#number of violations per boro
filtered['VIOLATION DESCRIPTION'].groupby(df['BORO']).count()

#most common
mostCommonViolationsPerBoro = filtered['VIOLATION DESCRIPTION'].groupby(df['BORO']).value_counts().groupby(level = 0).nlargest(1)
mostCommonViolationsPerBoro
```

```python
#Q13 => Phone Areas Correspond to a single zip code
def getAreaCode(series):
    return series[0:3]

df['AREACODE'] = df['PHONE'].map(getAreaCode)
mask = df['ZIPCODE'].groupby(df['AREACODE']).count() == 1

#number of Phone Areas that correspond to a single zipcode
len(df['ZIPCODE'].groupby(df['AREACODE']).count()[mask])
```

```python
def findLastWord(series):
    return series.split(" ")[-1]
def findBase(series):
    l = series.split(" ")
    while "" in l:
        l.remove("")
    l = l[0:len(l)-1]
    st = ""
    while len(l) != 0:
        st += str(l.pop()) + " "
    return st
df['STREET TYPE'] = df['STREET'].map(findLastWord)
df['STREET BASE'] = df['STREET'].map(findBase)
df['STREET BASE & ZIP'] = df['STREET BASE'] + " " + df['ZIPCODE']


newTbl = df[['STREET TYPE','STREET BASE','STREET BASE & ZIP']]
#remove duplicated
newTbl = newTbl.drop_duplicates()


newTbl
#drop Street base empty
mask = (newTbl['STREET BASE'] != "") | (newTbl['STREET BASE'] != " ")
newTbl = newTbl[mask]


#merge on it's self
newTbl = newTbl.merge(newTbl, left_on = 'STREET BASE & ZIP'
                      ,right_on = 'STREET BASE & ZIP',
                      suffixes = ['_left','_right'])
unequal = newTbl[newTbl['STREET TYPE_left'] != newTbl['STREET TYPE_right']]
ct = pd.crosstab(unequal['STREET TYPE_left'],unequal['STREET TYPE_right'])
IncorrectList = ['AVE','ST','RD','PL','BOULEARD','BULEVARD']


ct[IncorrectList].idxmax()
```
