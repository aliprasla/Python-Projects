## Question 13

To get the Phone Areas corresonding to a single zip code, we first extracted the  Area Code from the phone number, using _getAreaCode_. Then, we created a mask that filtered for the Area Codes that only contained a single Area Code. Finally, we returned the length of the filtered out series that contained the zipcodes.

```python
#Q13 => Phone Areas Correspond to a single zip code
def getAreaCode(series):
    return series[0:3]

df['AREACODE'] = df['PHONE'].map(getAreaCode)

mask = df[['AREACODE','ZIPCODE']].drop_duplicates()['ZIPCODE'].groupby(df['AREACODE']).count() == 1



#number of Phone Areas that correspond to a single zipcode
len(df['ZIPCODE'].groupby(df['AREACODE']).count()[mask])
```

## Question 14

First, we used map to find the last word ( findLastWord function) and the street base (findBase) of the streets in the dataset.

Next, we concatenated the base and zipcode, dropped the duplicate values, and merged the new table on to itself. Then, we applied a mask to determine which street types were misspelled. After applying a crosstab to determine frequencies for potential spellings of the base, we used _idxmax()_ to find the correct spelling.

```python
def findLastWord(series):
    return series.split(" ")[-1]
def findBase(series):
    l = series.split(" ")
    while "" in l:
        l.remove("")
    l = l[0:len(l)-1]
    st = ""
    while len(l) != 0:
        st += str(l.pop()) + " "
    return st


df['STREET TYPE'] = df['STREET'].map(findLastWord)
df['STREET BASE'] = df['STREET'].map(findBase)


df['STREET BASE & ZIP'] = df['STREET BASE'] + " " + df['ZIPCODE']


newTbl = df[['STREET TYPE','STREET BASE','STREET BASE & ZIP']]
#remove duplicated
newTbl = newTbl.drop_duplicates()


#merge on it's self
newTbl = newTbl.merge(newTbl, left_on = 'STREET BASE & ZIP'
                      ,right_on = 'STREET BASE & ZIP',
                      suffixes = ['_left','_right'])
#print newTbl

unequal = newTbl[newTbl['STREET TYPE_left'] != newTbl['STREET TYPE_right']]


ct = pd.crosstab(unequal['STREET TYPE_left'],unequal['STREET TYPE_right'])

#drop empty column
ct = ct[1:]

IncorrectList = ['AVE','ST','RD','PL','BOULEARD','BULEVARD']


ct[IncorrectList].idxmax()
```
